# Evaluación de modelos de regresión con scikit-learn

##Objetivo del ejercicio
Entrenar y evaluar un modelo de regresión sobre un conjunto de datos sintético, interpretando los resultados a partir de métricas comunes de desempeño como MAE, RMSE y R².

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## Paso 1: generar un conjunto de datos de regresión simulado

```python
# Generar un conjunto de datos con ruido
X, y = make_regression(n_samples=200, n_features=1, noise=15, random_state=42)

# Visualizar los datos
plt.scatter(X, y, color='blue', alpha=0.5)
plt.title('Datos simulados para regresión')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## Paso 2: dividir los datos en entrenamiento y prueba

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Paso 3: entrenar un modelo de regresión lineal

```python
model = LinearRegression()
model.fit(X_train, y_train)

# Predicciones sobre el conjunto de prueba
y_pred = model.predict(X_test)
```

## Paso 4: calcular métricas de evaluación

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Error absoluto medio (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Raíz del error cuadrático medio (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Coeficiente de determinación (R²)
r2 = r2_score(y_test, y_pred)

# Imprimir resultados
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.4f}")
```

## Interpretación de resultados esperados
- **MAE**: promedio de los errores absolutos entre las predicciones y los valores reales. Indica el margen de error promedio.
- **RMSE**: penaliza más fuertemente los errores grandes. Si el RMSE es significativamente mayor que el MAE, puede haber valores atípicos influyentes.
- **R²**: mide qué proporción de la varianza en la variable dependiente es explicada por el modelo. Un valor cercano a 1 indica buen ajuste.
